## D. Feature Engineering

```python
# Do not modify this code
try:
  df_eng = df_clean.copy()
except Exception as e:
  print(e)
```

### D.1 New Feature "CategoryName length"

```python
df_eng['CategoryName_Length'] = df_eng['CategoryName'].str.len()

df_eng['HasMultipleWords'] = df_eng['CategoryName'].str.contains(' ').astype(int)

print("New features created:")
print("CategoryName_Length:", df_eng['CategoryName_Length'].tolist())
print("HasMultipleWords:", df_eng['HasMultipleWords'].tolist())
print("\nDataset with new features:")
display(df_eng)
```

```python
# <Student to fill this section>
feature_engineering_1_explanations = """
New Features Created: CategoryName_Length and HasMultipleWords
Purpose: These are demonstration features showing feature engineering techniques. Category name length and word count patterns might reveal insights in larger datasets with more textual data.
Value: In real customer segmentation, we would create features like customer purchase frequency per category, spending patterns, or category preference scores. These examples illustrate the process of deriving new information from existing data.
Application: When joined with sales data, similar techniques can create customer-level features for meaningful clustering analysis.
"""
```

```python
# Do not modify this code
print_tile(size="h3", key='feature_engineering_1_explanations', value=feature_engineering_1_explanations)
```

### D.2 New Feature "Category type based on business logic"


```python
def categorize_product_type(category_name):
    if category_name == 'Bikes':
        return 'Primary_Product'
    elif category_name in ['Components', 'Accessories']:
        return 'Support_Product'
    elif category_name == 'Clothing':
        return 'Apparel'
    else:
        return 'Other'

df_eng['Category_Type'] = df_eng['CategoryName'].apply(categorize_product_type)

core_categories = ['Bikes', 'Components']
df_eng['IsCoreCategory'] = df_eng['CategoryName'].isin(core_categories).astype(int)

print("New business logic features:")
print("Category_Type:", df_eng['Category_Type'].tolist())
print("IsCoreCategory:", df_eng['IsCoreCategory'].tolist())
print("\nDataset with all engineered features:")
display(df_eng)
```

```python
feature_engineering_2_explanations = """
New Features Created: Category_Type and IsCoreCategory
Purpose: These features apply business logic to categorize products. Category_Type groups products by their business function, while IsCoreCategory identifies primary revenue-generating categories.
Business Value: In customer segmentation, these categories help identify customer preferences - whether they focus on core products, support items, or apparel. This enables targeted marketing strategies.
Real Application: When combined with customer purchase data, we can analyze which customer segments prefer which product types and adjust inventory and marketing accordingly.
"""
```

```python
# Do not modify this code
print_tile(size="h3", key='feature_engineering_2_explanations', value=feature_engineering_2_explanations)
```

### D.3 New Feature "Length_Score and Composite_Score"



```python
max_length = df_eng['CategoryName_Length'].max()
df_eng['Length_Score'] = df_eng['CategoryName_Length'] / max_length

df_eng['Composite_Score'] = (
    df_eng['Length_Score'] * 0.3 +
    df_eng['IsCoreCategory'] * 0.7
)

print("Final engineered features for clustering preparation:")
print("Length_Score:", df_eng['Length_Score'].round(3).tolist())
print("Composite_Score:", df_eng['Composite_Score'].round(3).tolist())

print("\nFinal engineered dataset:")
display(df_eng)

print("\nDataset info for clustering readiness:")
print(f"Shape: {df_eng.shape}")
print(f"Numerical features available: {df_eng.select_dtypes(include=['number']).columns.tolist()}")
```

```python
feature_engineering_n_explanations = """
New Features Created: Length_Score and Composite_Score
Purpose: These features demonstrate normalization and feature combination techniques essential for clustering algorithms. Length_Score normalizes text length to 0-1 scale, while Composite_Score combines multiple features with weighted importance.
Clustering Preparation: In real customer segmentation, similar techniques are used to create normalized, comparable features from diverse data sources (purchase frequency, recency, monetary value, etc.).
Business Impact: Properly engineered features enable meaningful distance calculations in clustering algorithms, leading to accurate customer segments that reflect real business patterns and behaviors.
"""
```

```python
# Do not modify this code
print_tile(size="h3", key='feature_engineering_n_explanations', value=feature_engineering_n_explanations)
```
